```python
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_counties_in_Ohio'
```

```python
tables = pd.read_html(url)
tables
```

```python
cnty_df = tables[0]
cnty_df.columns = ['County', 'FIPS Code', 'County Seat', 'Est.', 'Origin', 'Etymology', 'Population', 'Area', 'Map']
cnty_df.head()
```

```python
cnty_df = cnty_df.drop(['County Seat', 'Est.', 'Origin', 'Etymology', 'Area', 'Map'], axis=1)
cnty_df = cnty_df.rename(columns={'FIPS Code': 'County Code'})
cnty_df.head()
```

```python
cnty_df.County = cnty_df.County.str.replace('County','')
cnty_df.head()
```

```python
output_data_file = "resources/cnty_pop.csv"
```

```python
cnty_df.to_csv(output_data_file, index=False, header=True)
```
